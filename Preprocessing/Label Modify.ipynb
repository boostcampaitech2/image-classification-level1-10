### Label 수정하기

```python
df = pd.read_csv('/opt/ml/code/data/전체.csv')
df.head()
```

```python
gender_labeling_error = ['006359', '006360', '006361', '006362', '006363', '006364', '001498-1', '004432', '001720', '005223']
mask_labeling_error = ['000020', '004418', '005227']

df['gender_issue'] = df['id'].isin(gender_labeling_error)
df['mask_issue'] = df['id'].isin(mask_labeling_error) & df['stem'].isin(['normal', 'incorrect_mask'])

df.loc[(df['gender_issue'] == True) | (df['mask_issue'] == True)]
```

```python
df.loc[df['mask_issue'] == True, 'stem'] = df.loc[df['mask_issue'] == True, 'stem'].map({'normal':'incorrect_mask',
                                                                                         'incorrect_mask' : 'normal'})
df.loc[df['mask_issue'] == True, 'mask_issue'] = False
df.loc[df['gender_issue'] == True, 'gender'] = df.loc[df['gender_issue'] == True, 'gender'].map({'female' : 'male',
                                                                                                 'male' : 'female'})
df.loc[df['gender_issue'] == True, 'gender_issue'] = False
```

```python
print(df['gender_issue'].value_counts())
print(df['mask_issue'].value_counts())
```

```python
df.to_csv('/opt/ml/code/data/Final.csv', index = False)
```
